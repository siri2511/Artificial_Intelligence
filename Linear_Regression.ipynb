# Linear Regression



*   builds a model which establishes a relationship between features and targets
*   model has two parameters  ùë§ and ùëè whose values are 'fit' using training data.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x_train = input size in 1000 feet
# y_train = target price in 1000s dollars

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m = number of traning samples

m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
# (x^({i}), y^({i})) = i th training sample
# python indexing starts from 0

i = 0

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
i = 1

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# Scatter Plot
# arguments: (marker = x, c (color) = r)
#default color is blue

plt.scatter(x_train, y_train, marker= 'x', c= 'r')
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (in 1000 sqft)")
plt.show()
```

## Model Function

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b $$

```python
# w = weight
# b = bias

w = 100
b = 100
print(f"w : {w}")
print(f"b : {b}")
```

(ndarray (m,)) = numpy n-dimensional array of shape (m,) <br>
(scalar) = argument without dimensions, only magnitude <br>
np.zero(n) = one-dimensional numpy array with n entries <br>

```python
# model computation

def compute_model_output(x, w, b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b # model function

  return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w , b)

# model prediction
plt.plot(x_train, tmp_f_wb, c = 'b', label = 'Prediction')

# data points
plt.plot(x_train, y_train, marker = 'x', c = 'r', label = 'Actual Values')

plt.title("Housing Prices")
plt.ylabel("Price (in 1000s dollars)")
plt.xlabel("Size (in 1000 sqft)")
plt.legend()
plt.show()
```
